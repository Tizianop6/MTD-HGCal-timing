```python
import uproot
```

```python
file = uproot.open("step3.root")
```

```python
tree = file["Events"]
```

```python
tree.keys()
```

```python
l = [(b.name, b.compressed_bytes) for b in tree.values()]
```

```python
for name, size in l:
    if size > 0.1*tree.compressed_bytes:
        print(name, size/(2**(16)))
```
