```python
files = {ntuple : f"/grid_mnt/data_cms_upgrade/cuisset/supercls/alessandro_electrons/supercls-v14-dnn-Apr05_14-48/step3_DNN_{ntuple}.root" for ntuple in itertools.chain(range(1, 31), range(201,231))} #
```

```python
def processSample(ntuple, file:str):
    events = EdmReader(file)
    maker = SuperclsDataframeMaker()
    for evt in events:
        maker.processEvent(evt, ntuple)
    (Path(file).parent /  "dataframes").mkdir(exist_ok=True)
    maker.makeDf().to_pickle(Path(file).parent /  "dataframes" / f"supercls_{ntuple}.pkl.gz")
```

```python
with ProcessPoolExecutor(max_workers=15) as exc:
    exc.map(processSample, files.keys(), files.values())
```
