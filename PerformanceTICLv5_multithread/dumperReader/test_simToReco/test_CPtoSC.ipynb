```python
# aggregate tracksters in superclusters
supercls_grouped_df = (tsInSupercls_df
    .groupby(["eventInternal", "supercls_id"])
    .agg(**tracksters_groupBy(tsInSupercls_df, suffix="_supercls_sum"), caloparticle_id=pd.NamedAgg("caloparticle_id", "first"))
    .reset_index("supercls_id")
    .set_index("caloparticle_id", append=True)
)
supercls_grouped_df
```

```python
# merge with CaloParticle information
merged_CP = pd.merge(supercls_grouped_df, simTrackstersCP_df,
                how="left", right_index=True, left_index=True)
merged_CP.rename(columns={str(col) : str(col)+"_CP" for col in simTrackstersCP_df.columns})
```

```python
# merge with supercluster df to get supercls_id
# INDEX eventInternal	supercls_id, COLS simts_id	ts_id
merged_tsId = (pd.merge(df_mainAssoc, supercluster_all_df,
        how="left", 
        left_index=True, right_on=["eventInternal", "ts_id"]
    )
    .reset_index("ts_in_supercls_id", drop=True)
)
merged_tsId
```

```python
# merge with supercluster df to get supercls_id
# INDEX eventInternal	supercls_id, COLS genparticle_id	ts_id
merged_tsId = (pd.merge(df_mainAssoc, reader.supercluster_all_df,
        how="left", 
        left_index=True, right_on=["eventInternal", "ts_id"]
    )
    .reset_index("ts_in_supercls_id", drop=True)
)
merged_tsId
```

```python
# merge supercls_id to merged properties of superclusters
merged_supercls = pd.merge(
    merged_tsId, reader.supercluster_merged_properties_all, 
    left_index=True, right_index=True
).rename(columns={"ts_id" : "seed_ts_id"}).reset_index("supercls_id").set_index("genparticle_id", append=True)
merged_supercls
```

```python

# merge with CaloParticle information
merged_CP = pd.merge(merged_supercls, reader.simTrackstersCP_df,
                how="left", right_index=True, left_index=True)
merged_CP.rename(columns={str(col) : str(col)+"_CP" for col in reader.simTrackstersCP_df.columns})
```
